```python
# NumPy Speed Test
import time
import numpy as np

n=1000000
python_list=list(range(n))
numpy_array=np.arange(n)

start=time.time()
list_result=[]
for x in python_list:
    list_result.append(x*2)
end=time.time()
list_time=end-start
print("Time taken by Python list:",list_time)

start=time.time()
array_result=numpy_array*2
end=time.time()
array_time=end-start
print("Time taken by NumPy array:",array_time)
```

#####OBSERVATION:


1.  When I ran the program, NumPy gave the result faster than the Python list for the same operation.
2.   The list took more time because it processes each element one by one using a loop.


1.   NumPy was quicker since it performs the operation on the entire array at once.
