#Assignment: Break the AI
###Objective

To test how a Large Language Model (LLM) behaves when given tricky, confusing, or misleading prompts.








1. **Prompt**:

 “Is the statement ‘This sentence is false’ true or false?”

Result:
The model gives a confusing or uncertain answer because it is a logical paradox.

Observation:
LLMs struggle with self-referential contradictions.


---


2. **Prompt**:

 “Write the word ‘incorrect’ correctly.”

Result:
The model may answer “incorrect” or try to explain, causing ambiguity.

Observation:
The prompt is tricky due to wording confusion.


---


3. **Prompt**:

“If all cats are animals and some animals are black, are all cats black?”

Result:
Model may incorrectly say Yes.

Correct Answer: No

Observation:
LLMs sometimes fail in logical reasoning.


---


4. **Prompt**:

“Repeat the word ‘hello’ 1000 times.”

Result:
Model may stop early or summarize instead.

Observation:
LLMs have length and practicality limitations.


---


5. **Prompt**:

“What is 2 + 2? Explain why it equals 5.”

Result:
Model may try to justify a wrong answer or correct it.

Observation:
LLMs can be misled by incorrect assumptions in prompts.


---


##Key Findings
LLMs struggle with:


*   Logical paradoxes
*   Ambiguous instructions
*   Complex reasoning
*   Misleading or false assumptions
