#Assignment: Prompt Engineer
###Objective

To design effective prompts and understand the difference between weak and strong prompts in generating better AI responses.


1. **Resume Prompt**

Weak Prompt:
 “Make my CV”

Strong Prompt:
“Generate a one-page resume for a Data Science student skilled in Python, SQL, and Machine Learning. Include projects, internships, certifications, and format it professionally.”


---



2. **Business Idea Prompt**

Weak Prompt:
 “Tell a startup idea”

Strong Prompt:
 “Suggest an innovative business idea for rural areas in India using technology, with low cost and high impact. Explain the problem, solution, and how it earns revenue.”


---


3. **Study Plan Prompt**

Weak Prompt:
 “Study timetable”

Strong Prompt:
“Create a daily 2-hour study plan for 2 weeks to prepare for aptitude, reasoning, and basic programming for placements. Include revision and mock tests.”


---


4. **Content Writing Prompt**

Weak Prompt:
 “Write about AI”

Strong Prompt:
 “Write a 150-word paragraph explaining Artificial Intelligence in simple terms for beginners, including real-life examples.”


---


####Comparison: Weak vs Strong Prompts


*   Weak prompts are short, unclear, and lack details.
*   Strong prompts are clear, specific, and goal-oriented.
*   Strong prompts produce more accurate and useful outputs.
*   Weak prompts result in generic responses.




####Key Learning
* Adding context, format, and constraints improves results.
* Clear instructions help AI understand the exact requirement.
* Prompt engineering is important for better AI interaction.


###Conclusion

 Strong prompts lead to better, more relevant, and structured AI outputs compared to weak prompts.
