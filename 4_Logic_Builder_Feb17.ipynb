```python
 #Logic Builder: Print numbers 1–50 with Fizz/Buzz logic and count occurrences using loops and functions.
def fizz_buzz_count(start,end):
    fizz=0
    buzz=0
    fizzbuzz=0

    for i in range(start,end+1):
        if i%3==0 and i%5==0:
            print("FizzBuzz")
            fizzbuzz+=1
        elif i%3==0:
            print("Fizz")
            fizz+=1
        elif i%5==0:
            print("Buzz")
            buzz+=1
        else:
            print(i)

    return fizz,buzz,fizzbuzz



fizz_count,buzz_count,fizzbuzz_count=fizz_buzz_count(1,50)

print("\n---Count Summary---")
print("Fizz count:", fizz_count)
print("Buzz count:", buzz_count)
print("FizzBuzz count:", fizzbuzz_count)
```
