```python
#Image as Numbers
from google.colab import files
uploaded=files.upload()

import cv2
import numpy as np
from matplotlib import pyplot as plt

filename=list(uploaded.keys())[0]

# Read image
img=cv2.imread(filename)

# Show image
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("Image")
plt.axis("off")
plt.show()

# Shape
print("Image Shape:", img.shape)

# Proper channel detection
if len(img.shape)==2:
    print("Number of Channels: 1 (True Grayscale Image)")
else:
    # Check if actually grayscale stored in 3 channels
    if np.all(img[:, :, 0]==img[:, :, 1]) and np.all(img[:, :, 1]==img[:, :, 2]):
        print("Number of Channels: 1 ")
    else:
        print("Number of Channels: 3 (RGB/BGR Color Image)")

# Pixel values
print("\nTop-left 5x5 pixels:")
print(img[0:5, 0:5])
```

**Explanation: Image as Numbers**

An image in computer vision is not seen as a picture by the computer. Instead, it is stored as a matrix (array) of numbers. Each number represents the intensity or color information of a pixel.


---



**Image Shape**

The shape of an image tells its structure:

* Height → number of rows (pixels vertically)

* Width → number of columns (pixels horizontally)

* Channels → number of color layers

Example:

(512, 512, 3)

This means:

* 512 pixels height

* 512 pixels width

* 3 color channels


---


**Channels in Image**

*Color Image (RGB/BGR)*

* Has 3 channels

* Each pixel contains 3 values: Blue, Green, Red



*Format used in OpenCV is BGR


*Grayscale Image*
* Has 1 channel
* Each pixel contains only one value
* Value represents brightness intensity


---



**Pixel Values**

Each pixel in an image is stored as numbers:

* In color images → 3 values per pixel
* In grayscale images → 1 value per pixel

A small part of an image can be represented like a matrix of numbers.



---


**Conclusion**

An image is a numerical representation where:

* Shape defines structure
* Channels define color information
* Pixel values define intensity and color

Thus, image processing is essentially mathematical manipulation of numbers.
