```python
#Storytelling with Graphs
import matplotlib.pyplot as plt


apps=["Instagram","YouTube","WhatsApp","Netflix","Others"]
hours = [3,2.5,1.5,2,1]

# bar chart
plt.figure()
plt.bar(apps,hours)
plt.title("Daily Screen Time Usage")
plt.xlabel("Apps")
plt.ylabel("Hours")
plt.show()

# pie chart
plt.figure()
plt.pie(hours,labels=apps,autopct="%1.1f%%")
plt.title("Screen Time Distribution")
plt.show()

# histogram
plt.figure()
plt.hist(hours)
plt.title("Screen Time Range")
plt.xlabel("Hours")
plt.ylabel("Frequency")
plt.show()
```

###**data story**
The graphs show how time is spent on different apps during the day. From the bar chart, it is clear that more time is spent on Instagram and YouTube compared to other apps. The pie chart also shows that these apps take up a major portion of total screen time. The histogram indicates that most app usage falls between 1 to 3 hours, showing moderate usage. Overall, the trend suggests that social media apps are used more frequently, while other apps are used less.
