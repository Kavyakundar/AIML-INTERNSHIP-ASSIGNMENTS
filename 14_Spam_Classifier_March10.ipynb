```python
#Spam Classifier Thinking
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# dataset link
url="https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/sms.tsv"

# load dataset
data=pd.read_csv(url,sep="\t",header=None,names=["label","message"])


data["label"]=data["label"].map({"ham":0,"spam":1})


X=data["message"]
y=data["label"]

vectorizer=CountVectorizer()
X=vectorizer.fit_transform(X)


model=MultinomialNB()
model.fit(X,y)


msg=input("Enter a message: ")
msg_vec=vectorizer.transform([msg])


prediction=model.predict(msg_vec)

if prediction[0]==1:
    print("Spam Message")
else:
    print("Not Spam Message")
```

###**Features**

*   The model considers the words present in each message as features.
*   It analyzes how frequently each word appears in the message.


*   Words like “free”, “win”, and “offer” help in identifying spam.
*   CountVectorizer converts text into numerical vectors for processing.


*   The model learns patterns of word usage from the training data.
###**Data Needed**

*   A dataset containing SMS or email messages is required.
*   Each message must be labeled as spam or not spam.


*   The dataset should include a variety of different messages.
*   A larger dataset helps improve model accuracy.


*   The data must be clean and properly formatted for training.
###**Possible Mistakes**

*   The model may fail to recognize new or unseen words.
*   It may classify normal messages as spam (false positive).


*   It may miss some spam messages (false negative).
*   Messages with mixed content can confuse the model.


*   The model relies on word frequency and does not understand actual meaning.
