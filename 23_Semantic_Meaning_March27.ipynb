##Assignment: Semantic Meaning

1. Happy – Joyful

 Both words express positive emotions and happiness. They are synonyms.

2. Big – Large

 Both indicate size and mean something greater in dimension.

3. Help – Assist

Both indicate providing support or aid to someone.

4. Easy – Simple

👉 Both mean not difficult to understand or do.

5. Angry – Furious

 Both express strong negative emotions, with furious being more intense.


##**Explanation**
Semantic similarity means how closely two words are related in meaning

Words with similar meanings (synonyms) have high semantic similarity.

It is used in:



*   Natural Language Processing (NLP)
*   Search engines
*   Chatbots
