```python
#Data Doctor
import pandas as pd

# creating a sample dataset
data=pd.DataFrame({
    "name":["Aadhi","Appu","Aadhi","Kavya",None],
    "age":[20,21,20,None,22],
    "city":["Mangalore","Bangalore","Mangalore","Delhi","Delhi"]
})

print("Original Data:")
print(data)

# check missing values
print("\nMissing values before cleaning:")
print(data.isnull().sum())

# fill missing values
data["age"]=data["age"].fillna(data["age"].mean())
data["name"]=data["name"].fillna("unknown")

# remove duplicates
data=data.drop_duplicates()

# standardize text (lowercase)
data["city"]=data["city"].str.lower()

print("\nCleaned Data:")
print(data)

print("\nMissing values after cleaning:")
print(data.isnull().sum())
```

#####**Data cleaning**--Data cleaning is an important step because the original dataset may contain missing values, duplicate entries, or inconsistent data. In this dataset, some values were missing and there were repeated records, which could lead to incorrect results during analysis. By filling missing values, we ensure that the dataset is complete and does not cause errors. Removing duplicates helps avoid repetition that can affect calculations. Standardizing text, such as converting city names to lowercase, makes the data consistent and easier to analyze. Overall, data cleaning improves the quality of the data and helps in getting accurate and meaningful results.
