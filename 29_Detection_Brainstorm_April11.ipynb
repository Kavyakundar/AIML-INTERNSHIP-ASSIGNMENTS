##Assignment: Detection Brainstorm
**Description**:

List 5 uses of face/object detection and design one solution.


---



✅ 5 Uses of Face/Object Detection

1. **Retail Stores (Customer Analytics)**

Object detection is used in malls and shops to track customer movement, analyze behavior, and improve product placement.

2. **Autonomous Vehicles**

Self-driving cars use object detection to identify pedestrians, vehicles, traffic signals, and obstacles for safe navigation.

3. **Online Exam Proctoring**

Face detection is used in online exams to ensure the correct student is present and to detect cheating or multiple faces.

4. **Agriculture Monitoring**

Object detection helps identify crop diseases, pests, and plant growth stages using drone or camera images.

5. **Smart Home Systems**

Used in smart doorbells and home security systems to detect visitors and recognize familiar faces.

###Designed Solution

**Problem Statement:**

In online examinations, students may cheat using multiple devices or by leaving the system unattended.

**Proposed Solution:**

Develop an AI-Based Online Exam Monitoring System using Face Detection.

**Working:**  
* Student logs into exam portal with webcam enabled
* System continuously detects face during exam
* If face is not detected → warning is shown
* If multiple faces are detected → exam is flagged
* Activity log is stored for review

**Technologies:**

* Python
* OpenCV
* Face Detection / Deep Learning models
* Web camera integration
* Cloud database

**Advantages:**
* Prevents cheating
* Ensures fair examination
* Real-time monitoring
* Easy automation for institutions

**Conclusion:**

Face and object detection improves automation, safety, and accuracy in many fields. The proposed system helps maintain academic integrity in online examinations.
