```python
#House Price Predictor
import numpy as np
from sklearn.linear_model import LinearRegression

plot_area=np.array([[400],[700],[900],[1100],[1400]])
cost=np.array([80,120,160,200,260])

model=LinearRegression()
model.fit(plot_area,cost)

print("Estimated Prices for Existing Data:")
for i in range(len(plot_area)):
    predicted=model.predict([plot_area[i]])
    print("Plot size:",plot_area[i][0],"sq ft -> Estimated price:",round(predicted[0],2),"thousand")

new_size=int(input("\nEnter plot size in sq ft: "))
predicted=model.predict([[new_size]])

print("Estimated price for",new_size,"sq ft plot:",round(predicted[0],2),"thousand")
```
