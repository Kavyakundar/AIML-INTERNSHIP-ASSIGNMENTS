```python
# Customer Segmentation
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


data = {
    "Annual_Income": [10, 12, 14, 20, 22, 45, 48, 50, 55, 58, 30, 33, 35, 38, 40],
    "Spending_Score": [20, 25, 15, 70, 75, 50, 55, 53, 60, 65, 30, 35, 40, 42, 45]
}

df = pd.DataFrame(data)


print("Top Rows of Dataset:")
print(df.head())


X = df[["Annual_Income", "Spending_Score"]]


kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(X)


print("\nDataset with Clusters:")
print(df)


plt.scatter(df["Annual_Income"], df["Spending_Score"], c=df["Cluster"])
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation")
plt.show()
```

After applying K-Means, customers are divided into 3 groups:

🔹 Cluster 1 – High Income, High Spending
These are premium customers
Spend more money
Important for business profit

🔹 Cluster 2 – Low Income, Low Spending
Budget customers
Less purchasing power
Lower contribution to revenue

🔹 Cluster 3 – Medium Income, Medium Spending
Average customers
Stable and regular buyers
