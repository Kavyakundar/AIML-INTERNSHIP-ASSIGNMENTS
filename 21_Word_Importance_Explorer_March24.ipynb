```python
#Word Importance Explorer
from sklearn.feature_extraction.text import TfidfVectorizer

documents=[
    "Education is the key to success and knowledge",
    "Technology is changing the world rapidly",
    "Health is wealth and fitness is important for life",
    "Artificial intelligence is the future of technology",
    "Books provide knowledge and improve education"
]

vectorizer=TfidfVectorizer(stop_words='english')
tfidf_matrix=vectorizer.fit_transform(documents)

words=vectorizer.get_feature_names_out()
tfidf_array=tfidf_matrix.toarray()

for i, doc in enumerate(tfidf_array):
    print(f"\nTop Keywords in Doc{i+1}:")

    top_indices=doc.argsort()[-3:][::-1]

    for index in top_indices:
        print(words[index])
```

#Explanation
###**TF-IDF**:
TF-IDF (Term Frequency–Inverse Document Frequency) is a technique used to find important words in a document compared to a group of documents.
###**How it works**:
*Term Frequency (TF)*:
Measures how often a word appears in a document

*Inverse Document Frequency (IDF)*:
Reduces the importance of common words appearing in many documents

*TF-IDF Score*:
Words with high scores are considered important keywords
###**Result Explanation**:
Each document has unique keywords identified using TF-IDF
Common words like “is”, “the”, “and” are removed

Words like:

education, knowledge → important in Document 1

technology → important in multiple tech-related documents

health, fitness → highlight the theme of Document 3
