```python
import nltk
from rake_nltk import Rake
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
text=input("Enter a paragraph: ")
```

```python
rake=Rake()

rake.extract_keywords_from_text(text)

keywords=rake.get_ranked_phrases()

print("Keywords:")
for keyword in keywords:
    print("-", keyword)
```
