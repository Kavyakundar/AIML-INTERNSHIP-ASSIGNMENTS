```python
#Movie Review Analyzer
from textblob import TextBlob

# 5 movie reviews
reviews=[
    "The movie was absolutely amazing and very entertaining",
    "I did not like the film, it was too slow and boring",
    "The movie was average with decent performances",
    "Fantastic storyline and great acting by the cast",
    "The plot was weak and the movie was disappointing"
]

# Sentiment analysis
for i, review in enumerate(reviews):
    analysis=TextBlob(review)
    polarity=analysis.sentiment.polarity

    print(f"\nReview {i+1}:{review}")

    if polarity>0.2:
        print("Sentiment:Positive")
    elif polarity<-0.2:
        print("Sentiment:Negative")
    else:
        print("Sentiment:Neutral")
```
