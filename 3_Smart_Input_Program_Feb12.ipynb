```python
#Smart Input Program
name=input("Enter your name: ")
age=int(input("Enter your age: "))
hobby=input("Enter your hobby: ")

if age<13:
    category="child"
elif age<20:
    category="teenager"
elif age<60:
    category="adult"
else:
    category="senior citizen"


print("\nHi",name+"!")
print("At",age,"years old,you are in the",category,"age group.")
print("Spending time on",hobby,"sounds really interesting!")
```
