```python
# KNN IN REAL LIFE

from sklearn.neighbors import NearestNeighbors
import numpy as np

data=np.array([
    [4, 2, 5],# User A
    [3, 2, 4],# User B
    [5, 4, 1] # User C
])

knn=NearestNeighbors(n_neighbors=2,metric='euclidean')
knn.fit(data)

# Find neighbors for User A
distances,indices=knn.kneighbors([data[0]])

print("Nearest Neighbors of User A:",indices)
print("Distances:",distances)
```

##**Netflix-like Recommendations Using KNN**

KNN (K-Nearest Neighbors) is a simple algorithm that finds items or users that are similar based on their behavior or ratings. On Netflix, for example:



*   Every user rates or watches movies/series.
*   KNN compares these ratings with other users to find the “nearest neighbors” — people with similar tastes.
*   Once similar users are identified, movies or shows that they liked but you haven’t watched yet are recommended to you.



In short, Netflix looks at people who watch and like the same content as you and suggests what they enjoyed. The more similar the preferences, the more accurate the recommendation.
####**Example**
Consider three users and their ratings for Action, Comedy, and Drama movies. User A likes Action and Drama the most, while User B has similar preferences. KNN identifies User B as the nearest neighbor to User A based on these ratings. Using this similarity, Netflix can recommend movies that User B liked but User A hasn’t watched yet. This way, the platform provides personalized suggestions by comparing users’ tastes and finding the most similar ones.
