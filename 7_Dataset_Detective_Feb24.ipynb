```python
#Dataset Detective
import pandas as pd

# loading iris dataset
data=pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")

# display top rows
print("Top rows:")
print(data.head())

# find column with highest value
max_col=data.max(numeric_only=True).idxmax()
print("\nColumn with highest value:", max_col)

# count missing values
print("\nMissing values:")
print(data.isnull().sum())
```

###Insights


1.   I used head() to view the first few rows and understand the structure of the dataset.
2.   The dataset contains flower measurements like sepal length, sepal width, petal length, and petal width.


1.   Among all columns, petal length or sepal length usually has higher values compared to others.
2.   There are no missing values in the dataset, so the data is clean and easy to use.


1.   Different species of flowers have different measurement patterns, which helps in classification.
